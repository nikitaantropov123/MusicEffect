# PROJECT 3: MUSIC INFLUENCE ON MENTAL HEALTH

   Many people nowadays are affected by a lot of kinds of mental illnensses, sometimes incurable. My project is based on survey, that took place in 2022 in US, Washington. Its goal was to research if music could help people that suffer from mental illnesses. Let's start researching.

```python
import pandas as pd
import seaborn as sb
import matplotlib as mlt
import matplotlib.pyplot as plt
import numpy as np
```

```python
main=pd.read_csv('data/mxmh_survey_results.csv')
```

```python
ind=pd.Index(range(1,737,1))
```

```python
main.set_index(ind,inplace=True)
```

```python
main.info()
```

```python
main['Instrumentalist']=main['Instrumentalist'].fillna('No Data')
```

```python
main.dropna(subset=['While working'],inplace=True)
```

```python
main['Composer']=main['Composer'].fillna('No Data')
```

```python
main['Music effects']=main['Music effects'].fillna('No Data')
```

```python
main.dropna(subset=['Age'],inplace=True)
```

```python
main.info()
```

```python
main.dropna(subset=['Primary streaming service'],inplace=True)
```

Lets start reseaching. Nowadays, 95% of music is played from streaming platforms. 

```python
count=main['Primary streaming service'].value_counts()
```

```python
sb.set_theme(style='darkgrid',
            palette='deep')
sb.displot(main['Primary streaming service'],aspect=2,
           height=8)
plt.title('Primary streaming service',
          fontdict={'size': 25})
plt.xlabel('Service', fontdict={'size': 25})
plt.ylabel('Quantity', fontdict={'size': 25})
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
```

Most of the parpicipators choose Spotify streaming service.

```python
sb.set_theme(style='darkgrid',
            palette='deep',)   
sb.displot(data=main,y='Age',x='Fav genre',aspect=3)
plt.xlabel('Genres', fontdict={'size': 25})
plt.ylabel('Age', fontdict={'size': 25}) 
```

As well, younger generation prefer pop and rock music. Older people will choose gospel or classiacl music.

```python
sb.set_theme(style='darkgrid',
            palette='deep')
sb.catplot(data=main,x='While working',y='Age')
```

As you can see here, most of the people are listening to music while working. Also, we can see, that most of them between 15 to 30 years old which is younger part of modern society.

```python
sb.scatterplot(data=main,x='Hours per day',y='Age')
```

If we will take a look on a graph above, we could see that younger people are listening to music more than older ones. 

```python
main['Hours per day'].mean()
```

The average of hours of listening to music through the survey is approx. 3,5 hours a day. 

```python
anx=main['Anxiety']
dep=main['Depression']
ins=main['Insomnia']
ocd=main['OCD']
age=main['Age']
hpd=main['Hours per day']
mus=main['Music effects']
sb.set_theme(style='darkgrid',
            palette='deep')
plt.xlabel('Age', fontdict={'size': 10})
plt.ylabel('Feeling mentally ill (1-10)', fontdict={'size': 15})
sb.kdeplot(y=anx,x=age)
sb.kdeplot(y=dep,x=age)
sb.kdeplot(y=ins,x=age)
sb.kdeplot(y=ocd,x=age)
sb.set_theme(style='darkgrid')
```

As we can see, the highest rating of feeling mentally ill were mostly given by 20 year old's and 60-70 year old's. Lets check how amount of hours of listening to music affects mental health.

```python
sb.set_theme(style='darkgrid',
            palette='deep')
plt.xlabel('Age', fontdict={'size': 15})
plt.ylabel('Hours of music per day', fontdict={'size': 15})
sb.kdeplot(y=hpd,x=age)
```

We can see almost the same pirture up here.

```python
sb.set_theme(style='darkgrid',
            palette='deep')
sb.catplot(y=mus,x=age)
plt.xlabel('Age', fontdict={'size': 15})
plt.ylabel('Music influnce', fontdict={'size': 15})
sb.displot(x=mus,y=age)
plt.xlabel('Music influnce', fontdict={'size': 15})
plt.ylabel('Age', fontdict={'size': 15})
```

Now, we see that music can impove the mental health at any age. Especially at the age of 20, where were the most of the problems.
